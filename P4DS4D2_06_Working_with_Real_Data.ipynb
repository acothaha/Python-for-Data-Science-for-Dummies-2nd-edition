# Uploading, Streaming, and Sampling Data 

## Uploading small amounts of data into memory 

```python
with open("dataset_toy\Colors.txt", 'r') as open_file:
    print(f'Colors.txt content:\n {open_file.read()}')
```

## Streaming large amounts of data into memory 

```python
with open("dataset_toy\Colors.txt", 'r') as open_file:
    for i in open_file:
        print(f'Reading data: {i}')
```
