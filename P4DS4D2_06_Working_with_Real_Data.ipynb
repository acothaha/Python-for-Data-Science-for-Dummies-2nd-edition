## Sampling data in different ways 

#### using mathematical operation to sampling the data

```python
n = 2
# assgining open() to open_file variable in a with syntax
with open("dataset_toy\Colors.txt", 'r') as open_file:
    # assigning i and j into the number and observation in open_file 
    # using for loop
    for i,j in enumerate(open_file):
        # makes only the even numbers appear
        if i % n == 0 and i != 0:
            print(f'Reading data: {str(i)} content: {j}')
```

#### using random() function from random module to sampling the data

```python
# importing random() func from random mod
# random() func generates value from 0 until 1
from random import random
# defining the sample size we want from the population
sample_size = 0.25
with open("dataset_toy\Colors.txt", 'r') as open_file:
    # assigning i and j into the number and observation in open_file 
    # using for loop
    for i, j in enumerate(open_file):
        # using random fuction to randomize the sample
        if random() <= sample_size:
            print(f'Reading data: {str(i)} content: {j}')
```

## Rendering the image

```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm
# assign an URL to a variable
example = ("https://upload.wikimedia.org//" +
               "wikipedia/commons/c/c9/Spb_06-2017_img42_Krestovsky_Stadium.jpg")
# using imread() to call the image from designated URL
# as_gray argument tells python to turn the color into grays cale
image = imread(example, as_gray=True)
# imshow() used to render the picture
plt.imshow(image, cmap=cm.gray)
# show() used to display the image
plt.show()
```

## Discovering image size and type 

```python
print(f"data type: {type(image)}, shape: {image.shape}")
```

## Cropping the image

```python
image2 = image[5:2000, 0:2000]
plt.imshow(image2, cmap=cm.gray)
plt.show()
```

## Resize the image

```python
image3 = resize(image2, (10000, 10000), mode='symmetric')
plt.imshow(image3, cmap=cm.gray)
print(f"data type: {type(image3)}, shape: {image3.shape}")
```

## Working with XML

```python
from lxml import objectify
import pandas as pd

# parsing the data file using objectify.parse() method
xml = objectify.parse(open('dataset_toy/XMLData.xml'))
# obtaining the root node using the getroot() method
root = xml.getroot()

# creating empty df with the correcct collumn
df = pd.DataFrame(columns=('Number', 'String', 'Boolean'))

# this for loop fills the dataframe 
# with four records from the XML file
# each in a <Record> node
for i in range(0, 4):
    # making a variable that contains all
    # the children for one <Record> node
    obj = root.getchildren()[i].getchildren()
    # The chilren earlier are loaded into a dict
    # with a specific key to match the df columns
    row = dict(zip(['Number', 'String', 'Boolean'], 
                   [obj[0].text, obj[1].text, 
                    obj[2].text]))
    # creates an actual row from the process earlier
    row_s = pd.Series(row)
    # give name to the row
    row_s.name = i
    # appends it to the df
    df = df.append(row_s)

print(df)
```
